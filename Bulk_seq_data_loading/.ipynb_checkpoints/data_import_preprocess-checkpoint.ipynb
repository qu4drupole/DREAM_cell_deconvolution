```python
#FUNCTION TO FIND WHICH ROWS IN SUPPLEMENTAL TABLE HAVE DATA
def linkRow(tag):
    return tag.name == 'tr' and any("http" in s for s in tag.strings)
```

```python
rawPattern = re.compile('(?i)raw|count')
#need to build this out to include more situations
otherPattern = re.compile('(?i)normalized')
normPattern = re.compile('(?i)tpm|fpkm|rpkm')
```

```python
#FUNCTION TO EXTRACT DOWNLOAD LINK
#what if multiple files have 'raw'?
#link locations seem different depending on the table
#Sometimes the links don't follow the convention! like here,
#  -- https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE107011
def linkTrapper(links,gse):
    #dlink_base = 'https://www.ncbi.nlm.nih.gov/geo/download/?acc='+gse+'&format=file'
    token1=token2=token3=0
    preferredLink = ['','','']
    if len(links) > 0:
        for l in links:
            if re.search(r'(?i)mirna',l):
                continue
            # LOG: make a note when there are multple options for data
            if normPattern.search(l):
                token1 = 3
                preferredLink[token1-1]=l
            elif otherPattern.search(l):
                token2 = 2
                preferredLink[token2-1]=l
            elif rawPattern.search(l):
                token3 = 1
                preferredLink[token3-1]=l
                if re.search('RAW\.tar',l):
                    break
    if any(t != 0 for t in [token1,token2,token3]):
        token = max(token1,token2,token3)
        dlink = preferredLink[token-1]
    elif len(links) == 1:
        dlink = links[0]
    else:
        dlink = 'null'
        print('No suitable link found for '+gse)
        print(links)
        print('\n')
    return dlink
```

### 1. DATA CRAWLER/PARSER

```python
#AGGREGATING DOWNLOAD LINKS FOR EACH SERIES
for gse in gseList['gse']:
    #print(gse+'\n')
    ncbi_address = 'https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc='+gse
    base_url = 'https://www.ncbi.nlm.nih.gov'
    with urllib.request.urlopen(ncbi_address) as test:
        soup = BeautifulSoup(test, 'html.parser')
        #Finding the table
        supToken = soup.find(string=re.compile("Supplementary")).parent
        for p in supToken.parents:
            if p.name == 'table':
                supTable = p
                break
        #Do some checkpoints--like, superseries
        
        #Find the link(s)
        links = []
        for t in supTable.find_all(linkRow):
            # check to make sure the file isn't just BED file
            if re.search('BED',t.contents[6].string) and not re.search(',',t.contents[6].string):
                continue
            else:
                links.append(t.next_element.string)
        #print(links)
        #what if multiple files have 'raw'?
        dlink_name = linkTrapper(links,gse)
        gseList.loc[gseList['gse']==gse,'down_fname'] = dlink_name
        if dlink_name != 'null':
            for row in supTable.find_all('tr'):
                if re.compile(dlink_name).search(str(row)):
                    hit_row = row.find(string=re.compile('http')).parent
                    dlink = hit_row['href']
                    gseList.loc[gseList['gse']==gse,'down_link'] = base_url+dlink
                    gseList.loc[gseList['gse']==gse,'down_size'] = row.contents[2].string
                    gseList.loc[gseList['gse']==gse,'down_ftype'] = row.contents[6].string
        else:
            gseList.loc[gseList['gse']==gse,'down_link'] = dlink_name
                    
```

```python
# FUNCTION TO READ IN DATA AND TEST IF FILE IS FORMATTED CORRECTLY
def formatCheck(fname,stopToken,gse,of):
    try:
        if re.search('STAR',fname):
            of.write('SKIPPED: ('+gse+') formatCheck(): stupid STAR format...\n')
            stopToken = 1
            return None, stopToken
        if len(os.listdir('./tmp/raw_output')) == 1:
            fname = os.listdir('./tmp/raw_output')[0]
        elif fname.endswith('.gz'):
            fname = re.sub('(\.\w*)\Z','',fname)
        try:
            if re.search('\.csv',fname):
                df=pd.read_csv('tmp/raw_output/'+fname)
            elif re.search ('\.txt|\.tsv',fname):
                df=pd.read_csv('tmp/raw_output/'+fname, sep='\t')
                if df.shape[1] <= 2:
                    df=pd.read_csv('tmp/raw_output/'+fname, sep='\s+')
            elif re.search ('\.xls',fname):
                # this may be an exception for how the files were extracted...
                if any(fname == f for f in os.listdir('./tmp')):
                    df=pd.read_excel('tmp/'+fname)
                else:
                    df=pd.read_excel('tmp/raw_output/'+fname)
            else:
                of.write('SKIPPED: ('+gse+') formatCheck(): Error reading count file\n')
                stopToken = 1
                return None, stopToken
        except:
            of.write('SKIPPED: ('+gse+') formatCheck(): Error reading count file\n')
            stopToken = 1
            return None, stopToken
        # check for index column as gene column
        if 'Unnamed: 0' in df.columns:
            if re.search('\.csv',fname):
                df=pd.read_csv('tmp/raw_output/'+fname, index_col=0)
            elif re.search ('\.txt|\.tsv',fname):
                df=pd.read_csv('tmp/raw_output/'+fname,sep='\t', index_col=0)
            elif re.search ('\.xls',fname):
                if any(fname == f for f in os.listdir('./tmp')):
                    df=pd.read_excel('tmp/'+fname, index_col=0)
                else:
                    df=pd.read_excel('tmp/raw_output/'+fname, index_col=0)
            df['gene_id'] = df.index.values
            df = pd.concat([df[['gene_id']], df.select_dtypes(exclude=['object'])], 1)

        # parsing FPKM tracking files and others
        if df.select_dtypes(include=['object']).shape[1] > 1:
            count_df = df.select_dtypes(include=['float','int64'])
            annot_df = df.select_dtypes(include=['object'])
            gene_col = list(filter(lambda x: isinstance(re.search('(?i)gene|symbol|name',str(x)),re.Match), annot_df.columns.to_numpy()))
            if len(gene_col) == 1:
                annot_df = annot_df[gene_col]
            elif any(re.search('(?i)symbol',col) for col in gene_col):
                gene_col = list(filter(lambda x: isinstance(re.search('(?i)symbol',x),re.Match), gene_col))
                if len(gene_col) == 1:
                    annot_df = annot_df[gene_col]
            else:
                of.write('SKIPPED: ('+gse+') formatCheck(): supplemental file too complicated\n')
                stopToken = 1
                return None, stopToken
            df = pd.concat([annot_df, count_df],1)

        if df.shape[1] < 2:
            of.write('SKIPPED: ('+gse+') formatCheck(): difficulty understanding file formatting\n')
            stopToken = 1
            return None, stopToken
    except:
        of.write('SKIPPED: ('+gse+') formatCheck(): unknown problem\n')
        stopToken = 1
        return None, stopToken
    return df, stopToken
```

```python
# FUNCTION TO GRAB THE NAME OF THE COUNTS COLUMN
def CountColumnName(df,stopToken=0):
    try:
        CountName = ''
        if not df.select_dtypes(include=['int']).empty:
            df_count = df.select_dtypes(include=['int'])
            if df_count.shape[1]==1:
                CountName = df_count.columns
            #need to include conditions for finding counts if multiple columns exist
        elif not df.select_dtypes(include=['float']).empty:
            df_count = df.select_dtypes(include=['float'])
            if df_count.shape[1]==1:
                CountName = df_count.columns
        else:
            print('Can\'t find count column')
            #Need some elegant way to return this
            stopToken=1
    except:
        print('Can\'t find count column')
        return None
        stopToken=1
    return CountName
```

```python
# FUNCTION TO GRAB THE NAME OF GENE NAME COLUMN
def GeneColumnName(df,stopToken,gse,of):
    try:
        GeneColName = ''
        if stopToken == 1:
            return None, None, stopToken
        df_text=df.select_dtypes(include=['object'])
        if df_text.shape[1] == 1:
            GeneColName = df_text.columns
        elif df_text.shape[1] > 1:
            #Add some parsing to find the genes
            of.write('SKIPPED: ('+gse+') GeneColumnName(): Gene annotations too complicated, too lazy to parse\n')
            stopToken = 1
            return None, None, stopToken
        else:
            of.write('SKIPPED: ('+gse+') GeneColumnName(): Gene annotations too complicated, too lazy to parse\n')
            stopToken = 1
            return None, None, stopToken
        # check if the gene names are split
        if re.search("\||_", df[GeneColName].sample(1).values[0][0]):
            gene_delim = re.search("\||_", df[GeneColName].sample(1).values[0][0]).group(0)
            df_annot = df[GeneColName[0]].str.split(re.escape(gene_delim),1,expand=True)
            col_test=[isinstance(re.search('ENSG',annot),re.Match) for annot in df_annot.sample(1).get_values()[0]]
            if any(col_test):
                true_col = col_test.index(True)
                df_annot.rename(columns={true_col:'scraped_ENSG_ID'}, inplace=True)
                GeneColName = df_annot.columns[list(map(lambda x: x == 'scraped_ENSG_ID', df_annot.columns))]
                if df.select_dtypes(exclude=['object']).shape[1]==1:
                    df_count = df.select_dtypes(exclude=['object'])
                    df = pd.concat([df_annot[GeneColName], df_count], 1)
                else:
                    of.write('SKIPPED: ('+gse+') GeneColumnName(): Can not find counts\n')
                    stopToken = 1
                    return None, None, stopToken
            else:
                of.write('SKIPPED: ('+gse+') GeneColumnName(): Weird format; could not find Ensemble IDs\n')
                stopToken = 1
                return None, None, stopToken
    except:
        of.write('SKIPPED: ('+gse+') GeneColumnName(): unknown problem \n')
        stopToken = 1
        return None, None, stopToken
    return GeneColName, df, stopToken
```

```python
# FUNCTION TO GET THE TYPE OF DATA NORMALIZATION
def getDataNormType(cds_df,gse_df,gse,b,gse_meta,of):
    try:
        d_fname = b[0]
        norm_type = 'unknown'
        if re.search('(?i)fpkm',d_fname):
            norm_type = 'FPKM'
        elif re.search('(?i)rpkm',d_fname):
            norm_type = 'RPKM'
        elif re.search('(?i)tpm',d_fname):
            norm_type = 'TPM'
        if norm_type == 'unknown':
            #try to figure out normalization from data, sometimes its in the 'gse_meta' object
            #e.g. do column sums == 1e6
            # LOG: I didn't determine the normalization type
            if 'data_processing' in gse_meta.phenotype_data.columns:
                proc_summary = gse_meta.phenotype_data['data_processing'][0]
                if re.search('(?i)fpkm',proc_summary):
                    norm_type = 'FPKM'
                elif re.search('(?i)rpkm',proc_summary):
                    norm_type = 'RPKM'
                elif re.search('(?i)tpm',proc_summary):
                    norm_type = 'TPM'
            # ADD IF STATEMENT TO DETECT IF COUNTS ARE INTEGERS->RAW
            # probably a better way to do this:
            elif all(isinstance(re.search('int',str(x)), re.Match) for x in np.random.choice(cds_df.select_dtypes(exclude=['object']).dtypes.to_numpy(),3)):
                norm_type = 'RAW'
            # check the column names
            elif any(re.search('(?i)fpkm',col_name) for col_name in cds_df.columns.to_numpy()):
                norm_type = 'FPKM'
            elif any(re.search('(?i)rpkm',col_name) for col_name in cds_df.columns.to_numpy()):
                norm_type = 'RPKM'
            elif any(re.search('(?i)tpm',col_name) for col_name in cds_df.columns.to_numpy()):
                norm_type = 'TPM'
        gse_df.loc[gse_df['gse']==gse,'data_normType'] = norm_type
    except:
        of.write('SKIPPED: ('+gse+') getDataNormType(): unknown problem \n')
        return None
    return gse_df
```

```python
# FUNCTION TO DOWNLOAD THE DATA FOR A GSE RECORD, 
# RETURNS A DICTIONARY THAT MATCHES SAMPLE IDs AND NCBI IDs, b, the GSE record, and stopToken
def DownloadData(gse, df, stopToken, of):
    try:
        if stopToken == 1:
            return None, None, None, stopToken
        #clear the download directory
        shutil.rmtree(r'tmp\raw_output', ignore_errors=True)
        #make a dictionary to store filenames with GSM names
        ncbi_id_dict = {}
        a=df.loc[df['gse']==gse,'down_link'].to_numpy()
        b=df.loc[df['gse']==gse,'down_fname'].to_numpy()
        gse_meta = GEOparse.get_GEO(gse)
        if re.search('SuperSeries',gse_meta.metadata['summary'][0]):
            of.write('SKIPPED: ('+gse+') GSE record is super series\n')
            stopToken = 1
            # LOG: record is superseries
            return None, None, None, stopToken
        if any(re.search('(?i)(single).(cell)|10x|chip-seq|atac-seq',metadata[0]) for metadata in gse_meta.metadata.values()):
            # LOG: record is single cell
            of.write('SKIPPED: ('+gse+') GSE record is not bulk transcriptomics\n')
            stopToken = 1
            return None, None, None, stopToken
        try:
            urllib.request.urlretrieve(a[0],'./tmp/'+b[0])
        except URLError:
            base='https://www.ncbi.nlm.nih.gov/geo/download/?acc='+gse+'&format=file&file='
            urllib.request.urlretrieve(base+b[0],'./tmp/'+b[0])
        except:
            of.write('SKIPPED: ('+gse+') download link did not work\n')
            stopToken = 1
            return None, None, None, stopToken
        if stopToken == 0:
            subprocess.call([r'C:\Program Files\7-Zip\7z','x',ntpath.join('tmp',b[0]),outdir])
            #finish unzipping the count files
            #Check for other compression types?
            if len(os.listdir('tmp/raw_output')) > 1:
                for f in os.listdir('tmp/raw_output'):
                    try:
                        ncbi_id = re.search('(GSM)[0-9]+',f).group(0)
                        subprocess.call([r'C:\Program Files\7-Zip\7z','x',ntpath.join('tmp','raw_output',f),outdir])
                        ncbi_id_dict[ncbi_id] = re.sub('(\.\w*)\Z','',f)
                    except AttributeError:
                        of.write('SKIPPED: ('+gse+') Just confusing\n')
                        stopToken = 1
                        return None, None, None, stopToken
    except:
        of.write('SKIPPED: ('+gse+') DownloadData(): unknown problem \n')
        stopToken = 1
        return None, None, None, stopToken
    return ncbi_id_dict, b, gse_meta, stopToken
```

```python
# FUNCTION TO CONVERT THE GENE NAMES PREP DATAFRAME; DROP 'NA' GENE SYMBOLS, SUM SAME GENE SYMBOLS
# THE INDIVIDUAL FILE TYPE NEED TO BE JOINED INTO A SINGLE DATAFRAME BY THIS POINT
# !!THIS FUNCTION IS VERY SLOW!!
def geneConverter(df, col_idx, stopToken, gse, of):
    try:
        if re.search('NM_|ENST|NR_^',df.loc[0,col_idx].values[0]):
            # LOG: can't convert gene ID format
            of.write('SKIPPED: ('+gse+') geneConverter(): Could not convert gene ID format\n')
            stopToken = 1
            return None, stopToken
        df['test_gid'] = df[col_idx.values[0]].apply(lambda x: re.sub('(\.\w*)\Z','',x)) #.values
        mg = mygene.MyGeneInfo()
        conv_df = mg.getgenes(df['test_gid'].to_numpy(),'symbol',as_dataframe=True,df_index=False) #to_dataframe=True
        conv_df.rename(columns={'symbol':'gene_symbol_sql'}, inplace=True)
        df=df.merge(conv_df[['query','gene_symbol_sql']],how='left',left_on='test_gid',right_on='query')

        gs_VC=df['gene_symbol_sql'].value_counts()
        for gs in gs_VC.loc[gs_VC>1].index:
            gs_idx=df.loc[df['gene_symbol_sql']==gs].index
            # CHECKPOINT: make sure we're dealing with floats as count type
            gs_sum=df.loc[gs_idx].select_dtypes(include=['float']).sum()
            df.loc[gs_idx[0],gs_sum.index.to_numpy()] = gs_sum
            df.drop(gs_idx[1:], inplace=True)
        df.dropna(inplace=True)
        # LOG: All the stats for gene conversion--how many rows were dropped? how many redundancies? etc?
    except:
        of.write('SKIPPED: ('+gse+') geneConverter(): Unknown problem \n')
        stopToken = 1
        return None, stopToken
    return df, stopToken
```

```python
# FUNCTION TO POP OUT RELEVENT CDS COLUMNS AND FORMAT TO IMPORT INTO DATABASE
# REMEMBER: Convert ncbi_id_dict to df (gse_sampleIDs) prior to running
# REMEMBER: Need to check the name column of sampleID_df
def popDataAndFormat(cds_df, gse, gse_df, sample_df, sampleID_df, stopToken, of):
    try:
        gseName = 'gse_'+gse
        normType = 'norm_'+gse_df[gse_df['gse']==gse]['data_normType'].values[0]
        #filter the sampleMeta by gse and (cell type!!)
        gse_sampleMeta = sample_df[sample_df['gse']==gse]
        #trim gse_sampleIDs to samples in sampleMeta
        if not all(cds_gsm in gse_sampleMeta['id'].values for cds_gsm in sampleID_df.index.values):
            gsm_idx = list(set(gse_sampleMeta['id'].values) & set(gse_sampleIDs.index.values))
            sampleID_df = sampleID_df.loc[gsm_idx]
            cds_df = cds_df[np.insert(sampleID_df.loc[gsm_idx]['source_name'].values,0,'gene_symbol_sql')]
        # add the cell.type info to the sampleID_df
        sampleID_df['cell.type'] = list(map(lambda x: gse_sampleMeta.loc[gse_sampleMeta['id']==x,'cell.type'].values[0] ,sampleID_df.index.to_numpy()))
        #get the cds column names that match the GSMs
        # This needs to be cleaned up...
        colname_idx = []        
        for target in sampleID_df['source_name']:
            cds_col = list(filter(lambda x: re.search(re.escape(x),target), cds_df.columns.to_numpy()))
            #this is a bit sloppy:
            if len(cds_col) == 0:
                cds_col = 'NA'
            elif len(cds_col) > 1:
                cds_col = cds_col[-1]
            else: 
                cds_col = cds_col[0]
            colname_idx.append(cds_col)
        if len(colname_idx) == 0:
            of.write('SKIPPED: ('+gse+') popDataAndFormat(): CDS does not include the GSMs mentioned in sampleMeta\n')
            stopToken = 1
            return None, None, stopToken
        sampleID_df['cds_colname'] = colname_idx
        #pop the relevent data
        colname_cds = list(filter(lambda x: x != 'NA', colname_idx))
        colname_cds.insert(0,'gene_symbol_sql')
        cds_df = cds_df[colname_cds]
        #build the new name for each column...
        for cname in cds_df.columns[1:].values:
            gsm_ID = 'gsm_'
            gsm_ID += sampleID_df[sampleID_df['cds_colname']==cname].index.values[0]
            fullName = '__'.join([gseName,gsm_ID,normType])
            cds_df.rename(columns={cname:fullName},inplace=True)
    except:
        of.write('SKIPPED: ('+gse+') popDataAndFormat(): unknown problem \n')
        stopToken = 1
        return None, None, stopToken
    return cds_df, sampleID_df, stopToken
```

```python
# FUNCTION TO CHECK READABILITY OF COUNT DATA
def dataCheck(f):
    if re.search('\.wig',f):
        return 1
    if re.search('\.bed',f):
        return 1
    try:
        if re.search('\.csv',f):
            df=pd.read_csv(f, comment='#')
            readToken = 1
        elif re.search('\.txt|\.tsv',f):
            df=pd.read_csv(f, sep='\t', comment='#')
            readToken = 1
        elif re.search('\.xls',f):
            df=pd.read_excel(f)
            readToken = 1
        else:
            # LOG: can't read it
            return 1
        if readToken == 1:
            if df.shape[1] == 1:
                # LOG: Issue parsing count data
                return 1
        #maybe check if the are still in a strange format, like excel...
    except:
        # LOG what went wrong
        return 1
    return 0
```

```python
# FUNCTION TO READ IN DATA FROM INDIVIDUAL SAMPLE FILES
def readFormat(f, stopToken, gse, of):
    try:
        stopToken = 0
        if re.search('STAR',f):
            of.write('SKIPPED: ('+gse+') Stupid STAR format\n')
            stopToken = 1
            return None, stopToken

        #read in various formats
        if re.search('\.csv',f):
            df=pd.read_csv(f, comment='#')
        elif re.search('\.txt|\.tsv',f):
            df=pd.read_csv(f, sep='\t', comment='#')
        elif re.search('\.xls',f):
            df=pd.read_excel(f)

        # in the event a gene is read as column name
        if any(isinstance(genePattern.search(col),re.Match) for col in df.select_dtypes(include=['object']).columns):
            of.write('SKIPPED: ('+gse+') readFormat(): File header appears to be missing \n')
            stopToken = 1
            return None, stopToken

        if df.select_dtypes(include=['object']).shape[1] > 1:
            # Is it an FPKM tracking file?
            if 'FPKM' in df.columns:
                count_df = df.select_dtypes(include=['float'])
                # Based on 1 sample I looked at...
                count_df = count_df[['FPKM']]
                count_df.rename(columns={'FPKM':re.match('(tmp/raw_output/)(.*)',f).group(2)+'_FPKM'}, inplace=True)
                annot_df = df.select_dtypes(include=['object'])
                gene_col = list(filter(lambda x: isinstance(re.search('gene',str(x)),re.Match), annot_df.columns.to_numpy()))
                if len(gene_col) == 1:
                    annot_df = annot_df[gene_col]
                    df = pd.concat([annot_df, count_df],1)
                    gs_VC=df[gene_col[0]].value_counts()
                    for gs in gs_VC.loc[gs_VC>1].index:
                        gs_idx=df.loc[df[gene_col[0]]==gs].index
                        # CHECKPOINT: make sure we're dealing with floats as count type
                        gs_sum=df.loc[gs_idx].select_dtypes(include=['float']).sum()
                        df.loc[gs_idx[0],gs_sum.index.to_numpy()] = gs_sum
                        df.drop(gs_idx[1:], inplace=True)
                elif 'gene_short_name' in gene_col:
                    annot_df = annot_df[['gene_short_name']]
                    df = pd.concat([annot_df, count_df],1)
                    # do a duplicate check first
                    gs_VC=df['gene_short_name'].value_counts()
                    for gs in gs_VC.loc[gs_VC>1].index:
                        gs_idx=df.loc[df['gene_short_name']==gs].index
                        # CHECKPOINT: make sure we're dealing with floats as count type
                        gs_sum=df.loc[gs_idx].select_dtypes(include=['float']).sum()
                        df.loc[gs_idx[0],gs_sum.index.to_numpy()] = gs_sum
                        df.drop(gs_idx[1:], inplace=True)
                else:
                    of.write('SKIPPED: ('+gse+') readFormat(): Can not parse gene IDs\n')
                    stopToken = 1
                    return None, stopToken
            # Accounting for other weird file types/tracking files
            elif re.search('(?i)count',str(df.columns)):
                # this seems too specific...
                count_col = list(map(lambda x: isinstance(re.search('(?i)gene',x),re.Match), df.columns.to_numpy()))
                if sum(count_col) == 1:
                    count_col = df.columns[count_col]
                    count_df = df[count_col]
                    count_df.rename(columns={count_col:re.match('(tmp/raw_output/)(.*)',f).group(2)}, inplace=True)
                annot_df = df.select_dtypes(include=['object'])
                # if ensembl gene ids are availble...
                gene_col = list(filter(lambda x: isinstance(re.search('ensembl',str(x)),re.Match), annot_df.columns.to_numpy()))
                if len(gene_col) == 1:
                    annot_df = annot_df[gene_col]
                    df = pd.concat([annot_df, count_df],1)
                    gs_VC=df[gene_col[0]].value_counts()
                    for gs in gs_VC.loc[gs_VC>1].index:
                        gs_idx=df.loc[df[gene_col]==gs].index
                        gs_sum=df.loc[gs_idx].select_dtypes(include=['float']).sum()
                        df.loc[gs_idx[0],gs_sum.index.to_numpy()] = gs_sum
                        df.drop(gs_idx[1:], inplace=True)
                else:
                    of.write('SKIPPED: ('+gse+') readFormat(): too complicated\n')
                    stopToken = 1
                    return None, stopToken
            elif any(df.sample(1).apply(lambda x: isinstance(re.search('ENSG',str(x)), re.Match))):
                gene_col = df.columns[df.sample(1).apply(lambda x: isinstance(re.search('ENSG',str(x)), re.Match))]
                annot_df = df[gene_col]
                count_df = df[[df.columns[-1]]]
                count_df.rename(columns={df.columns[-1]:re.match('(tmp/raw_output/)(.*)',f).group(2)}, inplace=True)
                df = pd.concat([annot_df, count_df], 1)
                gs_VC=df[gene_col[0]].value_counts()
                for gs in gs_VC.loc[gs_VC>1].index:
                    gs_idx=df.loc[df[gene_col[0]]==gs].index
                    gs_sum=df.loc[gs_idx].select_dtypes(exclude=['object']).sum()
                    df.loc[gs_idx[0],gs_sum.index.to_numpy()] = gs_sum
                    df.drop(gs_idx[1:], inplace=True)
            elif re.search('(?i)gene',str(df.columns)):
                gene_col = list(map(lambda x: isinstance(re.search('(?i)gene',x),re.Match), df.columns.to_numpy()))
                gene_col = df.columns[gene_col]
                # praying that the count column is the last column...
                # Can maybe check if any column names are in the filename...
                count_df = df[[df.columns[-1]]]
                count_df.rename(columns={df.columns[-1]:re.match('(tmp/raw_output/)(.*)',f).group(2)}, inplace=True)
                if len(gene_col) == 1:
                    annot_df = df[gene_col]
                elif any(re.search('(?i)symbol',col_name) for col_name in gene_col):
                    gene_col = list(filter(lambda x: isinstance(re.search('(?i)symbol',x), re.Match), gene_col))
                    annot_df = df[gene_col]            
                df = pd.concat([annot_df, count_df], 1)
                gs_VC=df[gene_col[0]].value_counts()
                for gs in gs_VC.loc[gs_VC>1].index:
                    gs_idx=df.loc[df[gene_col[0]]==gs].index
                    gs_sum=df.loc[gs_idx].select_dtypes(exclude=['object']).sum()
                    df.loc[gs_idx[0],gs_sum.index.to_numpy()] = gs_sum
                    df.drop(gs_idx[1:], inplace=True) 
            else:
                of.write('SKIPPED: ('+gse+') readFormat(): too complicated\n')
                stopToken = 1
                return None, stopToken
        # for simple files that just have a gene name and some counts
        # maybe check that there are only two columns...
        elif df.select_dtypes(exclude=['object']).shape[1] == 1:
            count_col = df.select_dtypes(exclude=['object']).columns
            df.rename(columns={count_col[0]:re.match('(tmp/raw_output/)(.*)',f).group(2)}, inplace=True)
        else:
            of.write('SKIPPED: ('+gse+') readFormat(): Multiple numeric types found in file\n')
            stopToken = 1
            return None, stopToken
        df.dropna(inplace=True)
    except:
        of.write('SKIPPED: ('+gse+') readFormat(): unknown problem \n')
        stopToken = 1
        return None, stopToken
    return df, stopToken
```

```python
# FUNCTION TO CLEAN UP THE DATA SPACE
def clean_tmp_files(b):
    for f in os.listdir('./'):
        if '.soft' in f:
            try:
                os.remove(f)
            except:
                continue
    try:
        os.remove(ntpath.join('tmp',b[0]))
    except:
        # I don't know delete it later
        pass
```

```python
# FUNCTION TO BUILD gse_sampleIDs FROM CONCATENATED COUNT DATA
def makeGSEsampleIDs(df, ncbi_id_dict, gse_meta, stopToken, gse, of):
    try:
        if re.search('(GSM)[0-9]+',str(df.columns)):
            #This was just in case there were other weird columns, but it's probably fine?
            for col in df.columns:
                ncbi_id = re.search('(GSM)[0-9]+',col).group(0)
                ncbi_id_dict[ncbi_id] = re.sub('(\.\w*)\Z','',col)
            gse_sampleIDs = pd.DataFrame.from_dict(ncbi_id_dict,orient='index',columns=['source_name'])
            source_colnames = df.select_dtypes(exclude=['object']).columns.get_values()
            gse_sampleIDs['source_name'] = source_colnames
        else:
            # find which column in gse_meta could match the IDs; random column name
            # np.random.choice(cds.select_dtypes(exclude=['object']).columns,3) <-could be useful if things get more complicated
            quick_col = df.select_dtypes(include=['float','int64']).columns
            # pick a random count column
            ri = np.random.randint(0,len(quick_col),1)[0]
            col_idx = gse_meta.phenotype_data.columns[gse_meta.phenotype_data.apply(lambda x: quickMatch(x,df.columns[ri]), 0).values]
            #CHECKPOINT: Are there confusing (i.e. multiple) columns? add stopToken
            if len(col_idx) == 1:
                #this will work instead of a dictionary
                gse_sampleIDs = gse_meta.phenotype_data[col_idx]
                gse_sampleIDs.rename(columns={col_idx.values[0]:'source_name'},inplace=True)
            elif 'title' in col_idx:
                gse_sampleIDs = gse_meta.phenotype_data[['title']]
                gse_sampleIDs.rename(columns={'title':'source_name'}, inplace=True)
            else:
                of.write('SKIPPED: ('+gse+') makeGSEsampleIDs(): Unable to parse or identify GSM accessions\n')
                stopToken = 1
                return None, stopToken
        # Test if the sample is Human
        try:
            _col = df.select_dtypes(include=['float','int64']).columns[0]
            test_idx = gse_sampleIDs.loc[[isinstance(re.search(_col,x),re.Match) for x in gse_sampleIDs['source_name']]].index
            if gse_meta.phenotype_data['taxid_ch1'].loc[test_idx].values[0] != '9606':
                of.write('SKIPPED: ('+gse+') makeGSEsampleIDs(): Not human samples\n')
                stopToken = 1
                return None, stopToken
        except:
            pass
    except:
        of.write('SKIPPED: ('+gse+') makeGSEsampleIDs(): unknown problem \n')
        stopToken = 1
        return None, stopToken
    return gse_sampleIDs, stopToken
```

```python
# FUNCTION TO START NEW GSE RECORD IN LOG FILE
def logNewGSE(of,gse):
    of.write('--------------------------\n')
    of.write('GSENAME__'+gse+'\n')
    of.write('--------------------------\n')
```

```python
# FUNCTION TO WRITE DATAFRAME TO SQL DATABASE
def writeDataToSQL(sampleID_df, cds_df, gse, stopToken, of):
    # sampleID_df = gse_sampleIDs.copy()
    # cds=cds_2.copy()
    # get rid of spaces in cell.types
    sampleID_df['cell.type'] = sampleID_df['cell.type'].apply(lambda x: re.sub(' ','_',x)).values
    
    # last ditch checkpoint
    if cds_df.shape[0] > 200000:
        stopToken=1
        of.write('SKIPPED: ('+gse+') writeDataToSQL() Too many genes. Something went wrong preparing the final dataframe \n')
        return stopToken
    
    try:
    #add the new data
        for ct in sampleID_df['cell.type'].unique():
            #print(ct)
            ct_gsms = sampleID_df[sampleID_df['cell.type']==ct].index.values
            # Get the columns for the individual cell type, in this case PBMCs
            ct_col = list(filter(lambda x: x.split('gsm_')[1].split('__')[0] in ct_gsms, cds_df.columns[1:]))
            ct_col.insert(0,'gene_symbol_sql')
            ct_cds = cds_df[ct_col]
            if all(ct.lower() != cellType for cellType in engine.table_names()):
                ct_cds.to_sql(ct.lower(),con=engine, if_exists='replace', chunksize=2000)
            else:
                sql_df = pd.read_sql_table(ct.lower(), con=engine)
                max_size = sql_df.shape[0]
                cds_for_sql = sql_df.merge(ct_cds,how='outer',on='gene_symbol_sql')
                if sum(cds_for_sql.duplicated()) == 0:
                    cds_for_sql.drop(columns=['index'], inplace=True)
                    cds_for_sql.to_sql(ct.lower(),con=engine, if_exists='replace',chunksize=2000)
                else:
                    stopToken=1
                    of.write('SKIPPED: ('+gse+') writeDataToSQL() '+ct+' There is something strange about the gene symbols \n')
    except:
        stopToken = 1
        of.write('ERROR: ('+gse+') writeDataToSQL(): Something went wrong \n')
        return stopToken
    return stopToken
```
